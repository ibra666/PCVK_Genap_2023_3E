```python
# 3. Buatlah histogram citra seperti output histogram berikut berdasarkan flowchart di bawah ini : (Gunakan gambar lena.jpg)
# import gambar 
img_1 = cv.imread('/content/drive/MyDrive/Gambar/lena.jpg')
height, width, depth = np.shape(img_1)
names = np.arange(256)

#pemberian warna 
blue = [0]*256
green = [0]*256
red = [0]*256

#perhitungan warna 
for y in range(0,height):
  for x in range(0,width):
    blue[img_1[y][x][0]] += 1
    green[img_1[y][x][1]] += 1
    red[img_1[y][x][2]] += 1

#menampilkan histogram warna dimana dengan mengidentifikasi warna RGB yang ada dalam gambar lena 
#buatlah histogram penggunaan warna 
#menggunakan plot dimana untuk membuat warna tersebut menjadi bagiannya sendiri-sendiri 
#dari sublot tersebut dibuatlah tampilan 3 warna yaitu axs 0 sebagai red, 1= green dan 2=blue 
names = np.arange(256)
fig,  axs = plt.subplots(1, 3, figsize=[20,5], sharex=True, sharey=True)
fig.suptitle('Histogram RGB plot')
fig.text(0.05, 0.5, 'Jumlah Kemunculan', va='center', rotation='vertical')
fig.text(0.5, 0.04, 'Intensitas warna', ha='center')
axs[0].bar(names, red, color='red')
axs[1].bar(names, green, color='green')
axs[2].bar(names, blue, color='blue')
```

```python
# Tugas 4
# Membuat histogram citra menggunakan library yang dimiliki oleh NumPy yaitu 'histogram'
colors = ("b","g","r")
channel_ids = (0, 1, 2)

# Membuat Histogram Plot
plt.xlim([0, 256])
for channel_id, c in zip(channel_ids, colors):
  histogram, bin_edges = np.histogram(
      img_1[:, :, channel_id], bins=256, range =(0,256)
  )
  plt.plot(bin_edges[0:-1], histogram, color=c)

  plt.xlabel("Intensitas Warna")
  plt.ylabel("Jumlah kemunculan")

  plt.show()
```
